```python
import random
```

```python
#random datetime generator
def random_date(start, end):
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = randrange(int_delta)
    return start + timedelta(seconds=random_second)

from datetime import datetime
d1 = datetime.strptime('1/1/2019 10:30 PM', '%m/%d/%Y %I:%M %p')
d2 = datetime.strptime('1/30/2019 1:00 AM', '%m/%d/%Y %I:%M %p')
print(random_date(d1, d2))  
```

```python
#random_generator_bernoulli
def random_gen(string_1, string_2):
    return np.random.choice((string_1, string_2), 1)
```

```python
#random_generator for a given range of numbers
def random_number_gen(min_, max_):
    return random.randrange(min_, max_)
```

```python
#Append all the data we need to a list.
fake_data = []
for i in range(0, 50):
    index = 15 + i
    log_ts = random_date(d1,d2)
    from_user = random_number_gen(62221234567890, 63221234567890)
    to_user = random_number_gen(62221234567890, 63221234567890)
    type_ = random_gen('Transfer', 'Deposit')
    amount = random_number_gen(1000, 100000)
    fake_data.append((index, str(log_ts), str(from_user), str(to_user), type_[0], amount))
```
